```python
import pandas as pd
data=pd.read_csv('clean_twitter_data.csv')
```

```python
data.head()
```

```python
data1=data[:100]
data1
```

```python
import nltk
nltk.download('punkt')
tot=''
for i in data1.index:
  tot=tot+data1['Cleaned Text'][i]
  words=nltk.word_tokenize(tot)
  trigram=nltk.trigrams(words)
  freq = nltk.FreqDist(trigram)
  for k,v in freq.items():
      print(k,v)
```

```python
tot=''
for i in data1.index:
  tot=tot+data1['Cleaned Text'][i]
  words=nltk.word_tokenize(tot)
  bigram=nltk.bigrams(words)
  freq = nltk.FreqDist(bigram)
l=list(freq.keys())
i=0
for k,v in freq.items():
 
  prob= freq[k]/tot.count(l[i][0])
  
  print("probality of ",l[i], "is",prob)
  i+=1
```

```python
import spacy
d = pd.DataFrame(index=data1['User'])
nlp = spacy.load('en_core_web_sm')
for i in data1.index:
  np= nlp(data1['Cleaned Text'][i])
  
  for noun_chunk in np.noun_chunks:
    
    l=noun_chunk.text
    d[l]=''
    print(l)
    prob=l.count(l)/tot.count(l)
  d[l][i]=prob


```

# **Question 2: Undersand TF-IDF and Document representation**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
l=[]
for i in data1.index:
  l.append(data1['Cleaned Text'][i])

result=vectorizer.fit_transform(l)


print('\nWord indexes:')
print(vectorizer.vocabulary_)
print('\ntf-idf value:')
print(result)
print('\ntf-idf values in matrix form:')
print(result.toarray())
bow = pd.DataFrame(result.toarray(), columns = vectorizer.get_feature_names())
bow


```

```python
from sklearn.metrics.pairwise import cosine_similarity
bow1 = pd.DataFrame(cosine_similarity(bow, dense_output=True))
bow1
```
